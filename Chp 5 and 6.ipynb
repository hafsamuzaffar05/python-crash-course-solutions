```python

car = "toyota"
print(car == "toyota")  # True
print(car == "honda")   # False
alien_color = "green"
if alien_color == "green":
    print("You got 5 points!")

alien_color = "yellow"
if alien_color == "green":
    print("You got 5 points!")
else:
    print("You got 10 points!")
alien_color = "red"
if alien_color == "green":
    print("5 points")
elif alien_color == "yellow":
    print("10 points")
else:
    print("15 points")
age = 3
if age < 2:
    print("Baby")
elif age < 4:
    print("Toddler")
else:
    print("Kid")
favorite_fruit = "cherry"
if favorite_fruit == "cherry":
    print("You really like cherries!")
username = "admin"
if username == "admin":
    print("Hello admin!")
user_list = []
if len(user_list) == 0:
    print("No users found!")
old_name = "ali"
new_name = "Ali"
if new_name.lower() == old_name:
    print("This name is taken.")
number = 3
if number == 1:
    print("1st")
elif number == 2:
    print("2nd")
elif number == 3:
    print("3rd")
else:
    print(str(number) + "th")
```
